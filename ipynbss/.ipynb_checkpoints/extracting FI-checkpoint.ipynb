```python
import cv2
import os
import shutil  # For deleting the old folder

# Path to the input video
video_path = "./papers/newv2.mp4"  
cap = cv2.VideoCapture(video_path)

# Check if the video file opened successfully
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

print("Video loaded successfully.")

# Output folder for extracted frames
output_folder = "extracted_framesnewv2"

# Remove old folder if it exists
if os.path.exists(output_folder):
    shutil.rmtree(output_folder)  # Delete old extracted frames
os.makedirs(output_folder)

frame_count = 0

while True:
    ret, frame = cap.read() 
    if not ret:
        break  # Stop when video ends

    frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.png")
    cv2.imwrite(frame_filename, frame)
    
    if frame_count % 50 == 0:
        print(f" Extracted {frame_count} frames...")

    frame_count += 1 

# Release resources properly
cap.release()
cv2.destroyAllWindows()

print(f"Extracted {frame_count} frames and saved in '{output_folder}'")
```

```python
import cv2
import numpy as np
import os

# Create a directory for silhouette images
silhouette_folder = "silhouettesnewv2"
if not os.path.exists(silhouette_folder):
    os.makedirs(silhouette_folder)

# Read all frames from the extracted frames folder
frame_files = sorted([f for f in os.listdir(output_folder) if f.endswith('.png')])  # Ensure only image files

if len(frame_files) == 0:
    print("Error: No frames found in the extracted frames folder!")
    exit()

# Read the first frame as the background reference
background_path = os.path.join(output_folder, frame_files[0])
background = cv2.imread(background_path)

if background is None:
    print(f"Error: Could not read background frame: {background_path}")
    exit()

background_gray = cv2.cvtColor(background, cv2.COLOR_BGR2GRAY)

# Process each extracted frame
for frame_filename in frame_files:  
    frame_path = os.path.join(output_folder, frame_filename)
    frame = cv2.imread(frame_path)

    if frame is None:
        print(f" Warning: Skipping unreadable frame: {frame_filename}")
        continue

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Compute absolute difference
    diff = cv2.absdiff(background_gray, gray_frame)

    # Apply fixed threshold without morphological operations
    _, silhouette = cv2.threshold(diff, 28, 255, cv2.THRESH_BINARY)

    # Save the silhouette image
    silhouette_path = os.path.join(silhouette_folder, frame_filename)
    cv2.imwrite(silhouette_path, silhouette)

print(f"Silhouettes saved in '{silhouette_folder}'")
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
silhouette_folder = "silhouettesnewv2"
# Folder paths
silhouette_folder = os.path.join(silhouette_folder)
white_pixel_threshold = 1000  # Minimum white pixel count for a frame to be considered

# Function to count the number of white pixels in the lower half of the image
def count_white_pixels_lower_half(img):
    height, width = img.shape
    lower_half = img[height//2:, :]  # Extract the lower half
    return np.count_nonzero(lower_half)  # Count white pixels (non-zero pixels)

# Read all frame files
image_files = sorted(os.listdir(silhouette_folder))

# Extract white pixel count for each frame
frame_pixel_counts = []
frame_numbers = []  # Store extracted frame numbers

for frame_filename in image_files:
    frame_path = os.path.join(silhouette_folder, frame_filename)
    img = cv2.imread(frame_path, cv2.IMREAD_GRAYSCALE)

    # Extract the numeric frame number from filenames like "frame_0000.png"
    frame_number = int(''.join(filter(str.isdigit, frame_filename.split('.')[0])))  # Extract only digits
    
    # Count white pixels in the lower half
    white_pixel_count = count_white_pixels_lower_half(img)
    
    # Only store frames with enough white pixels
    if white_pixel_count >= white_pixel_threshold:
        frame_pixel_counts.append(white_pixel_count)
        frame_numbers.append(frame_number)

# Exclude first and last 10% of frames before detecting valleys
num_frames = len(frame_pixel_counts)
start_idx = int(0.25 * num_frames)
end_idx = int(0.85 * num_frames)

filtered_frame_pixel_counts = frame_pixel_counts[start_idx:end_idx]
filtered_frame_numbers = frame_numbers[start_idx:end_idx]

# Find Valleys in White Pixel Counts
frame_pixel_counts_np = np.array(filtered_frame_pixel_counts)
valley_indices = argrelextrema(frame_pixel_counts_np, np.less, order=4)[0]  # Detect valleys

# Map detected valley indices to actual frame numbers
valley_frames = [filtered_frame_numbers[i] for i in valley_indices]
print("Detected Valleys at Frames:", valley_frames)

# Select First and Third Valley as Gait Cycle
if len(valley_frames) >= 3:
    first_valley_frame = valley_frames[0]
    third_valley_frame = valley_frames[2]
    print(f"Gait Cycle Identified: Frame {first_valley_frame} to {third_valley_frame}")
else:
    print("Not enough valleys detected to determine a gait cycle.")
    first_valley_frame, third_valley_frame = None, None

# Plot the White Pixel Counts and Valleys
plt.figure(figsize=(10, 5))
plt.plot(frame_numbers, frame_pixel_counts, label="White Pixel Count", color='black')
plt.scatter(valley_frames, [frame_pixel_counts[frame_numbers.index(f)] for f in valley_frames], 
            color='red', label="Detected Valleys")

# Mark first and third valley if they exist
if first_valley_frame and third_valley_frame:
    plt.axvline(x=first_valley_frame, color='green', linestyle='--', label=f"First Valley: {first_valley_frame}")
    plt.axvline(x=third_valley_frame, color='blue', linestyle='--', label=f"Third Valley: {third_valley_frame}")

# Plot styling
plt.xlabel("Frame Number")
plt.ylabel("White Pixel Count")
plt.title("White Pixel Counts and Valley Detection")
plt.legend()
plt.show()
```

```python
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Ensure 'first_valley_frame' and 'third_valley_frame' are defined before this step
if first_valley_frame is None or third_valley_frame is None:
    print(" Error: Gait cycle frames were not detected. Exiting GEI computation.")
    exit()

# **Load only the gait cycle frames**
image_paths = sorted(glob.glob('silhouettesnewv2/frame_*.png'))

if len(image_paths) == 0:
    print(" Error: No silhouettes found in 'silhouettesnewv2' folder.")
    exit()

# Ensure frame range is within valid limits
if first_valley_frame < 0 or third_valley_frame >= len(image_paths):
    print(f" Error: Frame range ({first_valley_frame}-{third_valley_frame}) is out of bounds.")
    exit()

image_paths = image_paths[first_valley_frame:third_valley_frame + 1]  # Ensure valid range

# Read the first frame to initialize GEI dimensions
im = cv2.imread(image_paths[0], cv2.IMREAD_GRAYSCALE)
if im is None:
    print(f" Error: Could not read the first image {image_paths[0]}")
    exit()

h, w = im.shape
black = np.zeros((h, w), np.uint8)
GEI = black.copy()

# **Function to find the center of mass of a silhouette**
def CenterOfMass(img):
    height, width = img.shape # Import height and width from frames
    X = [] # Sum of x-coordinates of each row
    Y = [] # Sum of y-coordinates of each column
    WoR = [] # Weights of rows
    WoC = [] # Weights of columns
    for i in range(height):
        Nx = [] # x-coordinates of silhouette pixels in a single row
        for j in range(width):
            if img[i,j] == 255: # If the pixel is white
                Nx.append(j) # Append its x-coordinate
        if len(Nx) != 0: # If the row is not empty
            X.append(sum(Nx)) # Append the sum of the x-coordinates of the row
            WoR.append(len(Nx)) # Append the number of white pixels of the row
    CoMX = sum(X) // sum(WoR) # The average of the x-coordinates is their center of mass on x

    for j in range(width):
        Ny = [] # y-coordinates of silhouette pixels in a single column
        for i in range(height):
            if img[i,j] == 255: # If the pixel is white
                Ny.append(i) # Append its y-coordinate
        if len(Ny) != 0: # If the column is not empty
            Y.append(sum(Ny)) # Append the sum of the y-coordinates of the column
            WoC.append(len(Ny)) # Append the number of white pixels of the column
    CoMY = sum(Y) // sum(WoC) # The average of the y-coordinates is their center of mass on y
    return CoMX, CoMY # Return the coordinates of the center of mass of the silhouette


#Finding the corner coordinates of the rectangle surrounding the silhouette
def Border(img):
    height, width = img.shape # Import height and width from frames
    indexl = [] # x-coordinates of left-most pixel of each row
    indexr = [] # x-coordinates of right-most pixel of each row
    for i in range(height): # For each row
        for j in range(width): # Start from the left-most pixel
            if img[i,j] != 0: # If the pixel is not black
                indexl.append(j) # Append its x-coordinate
                break # Move to next row
        for j in reversed(range(width)): # Start from the right-most pixel
            if img[i,j] != 0: # If the pixel is not black
                indexr.append(j) # Append its x-coordinate
                break # Move to next row
    x = min(indexl) # The left-most x-coordinate in all rows
    X = max(indexr) # The right-most x-coordinate in all rows
    indext = [] # y-coordinates of top-most pixel of each column
    indexb = [] # y-coordinates of bottom-most pixel of each column
    for j in range(width):
        for i in range(height): # Start from the top-most pixel
            if img[i,j] != 0: # If the pixel is not black
                indext.append(i) # Append its y-coordinate
                break # Move to next column
        for i in reversed(range(height)): # Start from the bottom-most pixel
            if img[i,j] != 0: # If the pixel is not black
                indexb.append(i) # Append its y-coordinate
                break # Move to next column
    y = min(indext) # The top-most y-coordinate in all columns
    Y = max(indexb) # The bottom-most y-coordinate in all columns
    return x,X,y,Y

#Cropping the silhouette based on the results from Border function
def Crop(img,x,X,y,Y):
    Crop = img[y:Y, x:X]
    return Crop

#Aligning the silhouettes based on their center of mass and averaging them
def Add(old,new,center,black,i): # Current GEI, new frame, center of mass, black background, frame index
    center = list(center) # Width and height of silhouettes
    height, width = old.shape # Height and width of the frames
    x = (width // 2) - center[0] # left x-coordinate of the silhouette's frame
    y = (height // 2) - center[1] # top y-coordinate of the silhouette's frame
    height, width = new.shape # Width and height of a black background
    X = (x + width) # right x-coordinate of the silhouette's frame
    Y = (y + height) # bottom y-coordinate of the silhouette's frame
    a = 1.0/(i + 1) # New silhouette weight
    b = 1.0 - a # Current GEI weight
    new = cv2.addWeighted(new, a, old[y:Y, x:X], b, 0.0) # Combine the new frame inside the silhouette
    old = cv2.addWeighted(black, a, old, b, 0.0) # Combine the new frame outside the silhouette
    old[y:Y, x:X] = new # GEI
    return old

#Importing the silhouettes, and getting their dimensions
path =image_paths
im = cv2.imread(path[0]) # Read the first frame
y,x,c = im.shape # Height, width and channels respectively

#Combining the functions
black = np.zeros((y,x), np.uint8) # Black background
GEI = black # Define GEI
i = 0 # Define frames index
for path in path: # For every frame in the path
    image = cv2.imread(path) # Read the frame
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) # Turn the frame to one channel
    x,X,y,Y = Border(image) # Find the borders of the silhouette
    cropped = Crop(image,x,X,y,Y) # Crop the silhouette
    center = CenterOfMass(cropped) # Find the center of mass of the silhouette
    GEI = Add(GEI,cropped,center,black,i) # Find the GEI with the latest frame
    i += 1 # Add 1 to index of frames

#Visualising the result
x,X,y,Y = Border(GEI) # Find the borders of the GEI
GEI = Crop(GEI,x,X,y,Y) # Crop the GEI
cv2.imwrite('Gait_Energy_Image.png', GEI) # Write the image
GEI = cv2.imread('Gait_Energy_Image.png') # Read the image
plt.imshow(GEI) # Show the image
plt.show()
print(" GEI saved as 'Gait_Energy_Image.png'")
```

```python
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt
from scipy.stats import entropy as scipy_entropy

# Ensure first_valley_frame and third_valley_frame are defined
if 'first_valley_frame' not in globals() or 'third_valley_frame' not in globals():
    raise ValueError("Error: Define 'first_valley_frame' and 'third_valley_frame' before running this cell.")

# Load silhouette images
image_paths = sorted(glob.glob('silhouettesnewv2/frame_*.png'))
if not image_paths:
    raise FileNotFoundError("Error: No silhouettes found in 'silhouettesnewv2' folder.")

# Validate frame indices
if first_valley_frame < 0 or third_valley_frame >= len(image_paths):
    raise IndexError(f"Error: Frame range ({first_valley_frame}-{third_valley_frame}) is out of bounds.")

image_paths = image_paths[first_valley_frame:third_valley_frame + 1]

# Read first image to determine dimensions
im = cv2.imread(image_paths[0], cv2.IMREAD_GRAYSCALE)
if im is None:
    raise IOError(f"Error: Could not read {image_paths[0]}")

h, w = im.shape
silhouettes = []

# Align silhouettes based on centroid shift
for path in image_paths:
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise IOError(f"Error: Could not read {path}")
    
    moments = cv2.moments(img)
    if moments['m00'] != 0:
        cx = int(moments['m10'] / moments['m00'])
        cy = int(moments['m01'] / moments['m00'])
        shift_x = (w // 2) - cx
        shift_y = (h // 2) - cy
        M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
        aligned = cv2.warpAffine(img, M, (w, h), borderValue=0)
        silhouettes.append(aligned)

# Convert to 3D array for entropy computation
silhouettes_array = np.stack(silhouettes, axis=2)  # h x w x num_frames
p_foreground = np.mean(silhouettes_array == 255, axis=2)
p_background = 1 - p_foreground

# Compute Gait Entropy Image (GEnI)
GEnI = np.zeros((h, w), dtype=np.float32)

for i in range(h):
    for j in range(w):
        probs = [p_foreground[i, j], p_background[i, j]]
        GEnI[i, j] = scipy_entropy(probs, base=2) if 0 < p_foreground[i, j] < 1 else 0

# Normalize GEnI to 0-255
GEnI = (GEnI / GEnI.max() * 255).astype(np.uint8)

# Crop to the region containing relevant gait information
x, X = np.where(GEnI.any(axis=0))[0][[0, -1]]
y, Y = np.where(GEnI.any(axis=1))[0][[0, -1]]
GEnI = GEnI[y:Y+1, x:X+1]

# Save and display the GEnI image
cv2.imwrite('Gait_Entropy_Image.png', GEnI)
plt.imshow(GEnI, cmap='gray')
plt.title("Gait Entropy Image")
plt.axis('off')
plt.show()

print("GEnI saved as 'Gait_Entropy_Image.png'")
```

```python
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Ensure gait cycle detection was done
if 'first_valley_frame' not in globals() or 'third_valley_frame' not in globals():
    print("Error: Run gait cycle detection cell first.")
    exit()

# Load silhouette images
image_paths = sorted(glob.glob('silhouettesnewv2/frame_*.png'))
if not image_paths:
    print("Error: No silhouettes found in 'silhouettesnewv2' folder.")
    exit()

# Ensure frame indices are valid
if first_valley_frame < 0 or third_valley_frame >= len(image_paths):
    print(f"Error: Frame range ({first_valley_frame}-{third_valley_frame}) is out of bounds.")
    exit()

# Select only the frames within the gait cycle
image_paths = image_paths[first_valley_frame:third_valley_frame + 1]
im = cv2.imread(image_paths[0], cv2.IMREAD_GRAYSCALE)
if im is None:
    print(f"Error: Could not read {image_paths[0]}")
    exit()

h, w = im.shape
silhouettes = []

# Align silhouettes to the center of mass
for path in image_paths:
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    moments = cv2.moments(img)
    if moments['m00'] != 0:
        cx = int(moments['m10'] / moments['m00'])
        cy = int(moments['m01'] / moments['m00'])
        shift_x = (w // 2) - cx
        shift_y = (h // 2) - cy
        M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
        aligned = cv2.warpAffine(img, M, (w, h), borderValue=0)
        silhouettes.append(aligned)

# Compute pixel-wise motion energy (Active Energy Image)
AEI = np.zeros((h, w), dtype=np.float32)

for i in range(1, len(silhouettes)):  # Start from second frame
    diff = np.abs(silhouettes[i] - silhouettes[i - 1])  # Frame difference
    AEI += diff

# Normalize AEI to range 0-255
AEI = (AEI / AEI.max() * 255).astype(np.uint8) if AEI.max() > 0 else AEI

# Crop to the region of interest
x, X = np.where(AEI.any(axis=0))[0][[0, -1]]
y, Y = np.where(AEI.any(axis=1))[0][[0, -1]]
AEI = AEI[y:Y+1, x:X+1]

# Save and display the AEI
cv2.imwrite('Active_Energy_Image.png', AEI)
plt.imshow(AEI, cmap='gray')
plt.title("Active Energy Image")
plt.show()

print("AEI saved as 'Active_Energy_Image.png'")
```

```python
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt
from scipy.stats import entropy as scipy_entropy

# Ensure first_valley_frame and third_valley_frame are defined
if 'first_valley_frame' not in globals() or 'third_valley_frame' not in globals():
    raise ValueError("Error: Run gait cycle detection first.")

# Load silhouette images
image_paths = sorted(glob.glob('silhouettesnewv2/frame_*.png'))
if not image_paths:
    raise FileNotFoundError("Error: No silhouettes found in 'silhouettesnewv2' folder.")

# Ensure frame indices are valid
if first_valley_frame < 0 or third_valley_frame >= len(image_paths):
    raise IndexError(f"Error: Frame range ({first_valley_frame}-{third_valley_frame}) is out of bounds.")

# Select only the frames within the gait cycle
image_paths = image_paths[first_valley_frame:third_valley_frame + 1]
im = cv2.imread(image_paths[0], cv2.IMREAD_GRAYSCALE)
if im is None:
    raise IOError(f"Error: Could not read {image_paths[0]}")

h, w = im.shape
silhouettes = []

# Align silhouettes based on the center of mass
for path in image_paths:
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    moments = cv2.moments(img)
    if moments['m00'] != 0:
        cx = int(moments['m10'] / moments['m00'])
        cy = int(moments['m01'] / moments['m00'])
        shift_x = (w // 2) - cx
        shift_y = (h // 2) - cy
        M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
        aligned = cv2.warpAffine(img, M, (w, h), borderValue=0)
        silhouettes.append(aligned)

# Convert silhouette list to a NumPy array for processing
silhouettes_array = np.stack(silhouettes, axis=2)  # h x w x num_frames

# Compute GEI (average silhouette image)
GEI = np.mean(silhouettes_array, axis=2).astype(np.uint8)

# Compute AEI (motion energy accumulation)
AEI = np.zeros((h, w), dtype=np.float32)
for i in range(1, len(silhouettes)):  # Start from second frame
    diff = np.abs(silhouettes[i] - silhouettes[i - 1])  # Frame difference
    AEI += diff

AEI = (AEI / AEI.max() * 255).astype(np.uint8) if AEI.max() > 0 else AEI

# Compute GEnI (Shannon entropy per pixel)
p_foreground = np.mean(silhouettes_array == 255, axis=2)
p_background = 1 - p_foreground
GEnI = np.zeros((h, w), dtype=np.float32)

for i in range(h):
    for j in range(w):
        probs = [p_foreground[i, j], p_background[i, j]]
        GEnI[i, j] = scipy_entropy(probs, base=2) if 0 < p_foreground[i, j] < 1 else 0

GEnI = (GEnI / GEnI.max() * 255).astype(np.uint8)

# Normalize and Crop each template
def crop_and_resize(img, target_shape):
    x, X = np.where(img.any(axis=0))[0][[0, -1]]
    y, Y = np.where(img.any(axis=1))[0][[0, -1]]
    cropped = img[y:Y+1, x:X+1]
    return cv2.resize(cropped, target_shape, interpolation=cv2.INTER_AREA)

# Set the reference shape (GEI shape after cropping)
GEI = crop_and_resize(GEI, (h, w))
AEI = crop_and_resize(AEI, (h, w))
GEnI = crop_and_resize(GEnI, (h, w))

# **Fix: Ensure FEI uses the correct final cropped shape**
final_h, final_w = GEI.shape  # Use cropped size as reference

# Create RGB Fusion Energy Image (FEI)
FEI = np.zeros((final_h, final_w, 3), dtype=np.uint8)

# Assign channels
FEI[..., 0] = AEI  # Red channel
FEI[..., 1] = GEI  # Green channel
FEI[..., 2] = GEnI  # Blue channel

# Save and display the FEI
cv2.imwrite('Fusion_Energy_Image.png', FEI)
plt.imshow(FEI)
plt.title("Fusion Energy Image (FEI)")
plt.show()

print("FEI saved as 'Fusion_Energy_Image.png'")
```
