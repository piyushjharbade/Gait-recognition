```python
import cv2
import os
import shutil

video_path = "./papers/newv2.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

print("Video loaded successfully.")
fps = cap.get(cv2.CAP_PROP_FPS)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
print(f"Video properties: FPS={fps}, Width={width}, Height={height}")

output_folder = "extracted_framesnewv2"
if os.path.exists(output_folder):
    try:
        shutil.rmtree(output_folder)
    except Exception as e:
        print(f"Error deleting folder: {e}")
        exit()
os.makedirs(output_folder)

frame_count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.png")
    if not cv2.imwrite(frame_filename, frame):
        print(f"Error saving frame {frame_count}")
        break
    
    if frame_count % 50 == 0:
        print(f"Extracted {frame_count + 1} frames...")

    frame_count += 1

cap.release()
cv2.destroyAllWindows()
print(f"Extracted {frame_count} frames and saved in '{output_folder}'")
```

```python
import cv2
import numpy as np
import os

silhouette_folder = "silhouettesnewv2"
if not os.path.exists(silhouette_folder):
    os.makedirs(silhouette_folder)

frame_files = sorted([f for f in os.listdir(output_folder) if f.endswith('.png')])
if not frame_files:
    print("Error: No frames found in the extracted frames folder!")
    exit()

subtractor = cv2.createBackgroundSubtractorMOG2(history=50, varThreshold=20, detectShadows=False)
kernel = np.ones((3,3), np.uint8)

for frame_filename in frame_files:
    frame_path = os.path.join(output_folder, frame_filename)
    frame = cv2.imread(frame_path)
    if frame is None:
        print(f"Warning: Skipping unreadable frame: {frame_filename}")
        continue

    silhouette = subtractor.apply(frame)
    silhouette = cv2.erode(silhouette, kernel, iterations=1)
    silhouette = cv2.dilate(silhouette, kernel, iterations=1)

    silhouette_path = os.path.join(silhouette_folder, frame_filename)
    if not cv2.imwrite(silhouette_path, silhouette):
        print(f"Error saving silhouette {frame_filename}")

print(f"Silhouettes saved in '{silhouette_folder}'")
```

```python
import numpy as np
import cv2
from scipy.signal import find_peaks
from matplotlib import pyplot as plt
import glob
silhouette_files = sorted(glob.glob('silhouettesnewv2/frame_*.png'))
centroids_y = []

for silhouette_path in silhouette_files:
    silhouette = cv2.imread(silhouette_path, cv2.IMREAD_GRAYSCALE)
    if silhouette is None:
        continue
    moments = cv2.moments(silhouette)
    if moments['m00'] != 0:
        cy = int(moments['m01'] / moments['m00'])
        centroids_y.append(cy)
    else:
        centroids_y.append(np.nan)

centroids_y = np.array(centroids_y)
centroids_y = np.nan_to_num(centroids_y, nan=np.nanmean(centroids_y))

smoothed_centroids = np.convolve(centroids_y, np.ones(5)/5, mode='valid')
valleys, _ = find_peaks(-smoothed_centroids, distance=10, prominence=10)

print(f"Detected Valleys at Frames: {valleys}")
if len(valleys) >= 3:
    first_valley_frame = valleys[0]
    third_valley_frame = valleys[2]
    print(f"Gait Cycle Identified: Frame {first_valley_frame} to {third_valley_frame}")
else:
    print("Error: Insufficient valleys detected. Using first detected cycle.")
    first_valley_frame = valleys[0] if len(valleys) > 0 else 0
    third_valley_frame = valleys[1] if len(valleys) > 1 else len(smoothed_centroids) - 1

plt.plot(smoothed_centroids)
plt.plot(valleys, smoothed_centroids[valleys], "x")
plt.title("Centroid Y Movement with Valleys")
plt.show()
```
